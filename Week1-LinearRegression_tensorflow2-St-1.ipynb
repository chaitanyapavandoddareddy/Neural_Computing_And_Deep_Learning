# Linear Regression Example using tensorflow
Linear regression implementation with TensorFlow v2 library.

This example is using a low-level approach to better understand all mechanics behind the training process.


```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
rng = np.random
```

```python
learning_rate = 0.01
training_steps = 10000
display_steps = 50
```

```python
# Just run the next cell using 'Ctrl' + 'Enter'
```

```python
# Training Data.

X = np.array([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
              7.042,10.791,5.313,7.997,5.654,9.27,3.1])

Y = np.array([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
              2.827,3.465,1.65,2.904,2.42,2.94,1.3])

n_samples = X.shape[0]
```

```python
print(X,'\n', X.shape)
```

```python
print(Y,'\n', Y.shape)
```

```python
W = tf.Variable(rng.randn(), name = "weights")
b = tf.Variable(rng.randn(), name = "bias")

print(W.numpy(), '\n', b.numpy())
```

```python
def linear_regression(x):
    return W*x +b
```

```python
def mean_square(y_pred, y_true):
    return tf.reduce_sum(tf.pow(y_pred-y_true, 2))/(2*n_samples)
```

```python
optimizer = tf.optimizers.SGD(learning_rate)
```

```python
# Adding a question mark () after the function name lets you view the latest documentation. 
# This is a handy tool.
```

```python
tf.optimizers.SGD?
```

```python
def run_optimization():
    
    with tf.GradientTape() as g:
        pred = linear_regression(X)
        loss = mean_square(pred,Y)
    
    gradients = g.gradient(loss, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))
```

```python
for step in range(1, training_steps +1):
    run_optimization()
    
    if step % display_steps == 0:
        pred = linear_regression(X)
        loss = mean_square(pred, Y)
        print ("step: %i, Loss: %f, W: %F, b %F" %(step, loss, W.numpy(), b.numpy()))
```

```python
plt.plot(X,Y, 'g^', label = 'Origional Data')
plt.plot(X,np.array(W * X + b), label = 'Fitted line')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
# Plot the original data points
plt.plot(X, Y, 'g^', label='Original Data', markersize=8)  # Increased marker size

# Plot the fitted line
plt.plot(X, np.array(W * X + b), label='Fitted Line', linewidth=2)  # Increased line width

# Add grid, title, and labels for clarity and aesthetics
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.title('Data Points and Fitted Line')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')

# Add a legend with a shadow
plt.legend(loc='best', shadow=True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have your X, Y, W, and b defined here

# Using a style template for a sophisticated look
plt.style.use('fivethirtyeight')

# Create a figure and a set of subplots
fig, ax1 = plt.subplots()

# Plotting the original data points on the first subplot
color = 'tab:green'
ax1.set_xlabel('X-axis label')
ax1.set_ylabel('Y-axis label', color=color)
ax1.plot(X, Y, 'o', color=color, label='Original Data')
ax1.tick_params(axis='y', labelcolor=color)

# Adding error bars to the data points
error = np.random.rand(*Y.shape) * 0.5  # Random error for demonstration
ax1.errorbar(X, Y, yerr=error, fmt='o', color=color, alpha=0.5)

# Create a secondary axis for the fitted line
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Fitted Line', color=color)  # we already handled the x-label with ax1
ax2.plot(X, np.array(W * X + b), color=color, linewidth=2, label='Fitted Line')
ax2.tick_params(axis='y', labelcolor=color)

# Adding a legend to the plot
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Adding an interactive widget (e.g., a slider) for demonstration
# Note: This requires an interactive backend like Jupyter Notebook.
from ipywidgets import interact, widgets

def update(w=1.0):
    ax2.clear()
    ax2.plot(X, np.array(w * X + b), color=color)
    fig.canvas.draw_idle()

interact(update, w=widgets.FloatSlider(value=1.0, min=0.5, max=2.0, step=0.1))

# Show the plot
plt.show()
```
